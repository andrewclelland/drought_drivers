```python
import ee
ee.Authenticate()
ee.Initialize(project='spherical-berm-323321')
```

```python
import geemap
```

```python
# Import shapefiles/images from assets
final = ee.FeatureCollection('users/andyc97/model_shapefiles/final_shapefile')
final_grid = ee.FeatureCollection('users/andyc97/model_shapefiles/final_grid')
tem_final = ee.FeatureCollection('users/andyc97/Shapefiles/TEM_Final_Domain')
tem_map = ee.Image('users/andyc97/model_shapefiles/TEM_LandCover_Map_V4')
```

```python
# Create palette for viewing image
mapVis = {'palette': ['ff0000', 'fd4100', 'fb8200', 'f9c400', 'f2ff00', 'b6ff05',
                     '7aff0a', '3eff0f', '02ff15', '00ff55', '00ff99', '00ffdd',
                     '00ddff', '0098ff', '0052ff', '0210ff', '3a0dfb', '7209f6',
                     'a905f1', 'e102ed', 'ff00cc', 'ff0089', 'ff0047', 'ff0004'],
         'min':0, 'max':99}
```

```python
# Create desired shapefile
# WWF All Boreal/Tundra Ecosystems
boreal = ee.FeatureCollection('users/andyc97/Shapefiles/wwf_ecosystems').filter('BIOME==6')
tundra = ee.FeatureCollection('users/andyc97/Shapefiles/wwf_ecosystems').filter('BIOME==11')
wwf = boreal.merge(tundra)
```

```python
# Example for loading country borders
countries = ee.FeatureCollection("FAO/GAUL/2015/level1")
norway = countries.filter(ee.Filter.eq("ADM0_NAME", "Norway"))
```

```python
# Export shapefile to asset
task = ee.batch.Export.table.toAsset(
    collection=wwf,
    description='boreal_tundra_shapefile',
    assetId='users/andyc97/model_preprocessed/boreal_tundra_shapefile',
    maxPixels=1e13)
task.start()
```

```python
# Visualise on map
Map = geemap.Map()
Map.addLayer(tem_map.updateMask(tem_map.neq(0)), mapVis, 'TEM Map')
Map.addLayer(final, {}, 'Final')
Map
```
